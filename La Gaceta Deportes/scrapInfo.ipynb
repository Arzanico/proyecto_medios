```python
import pandas as pd
import json
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
```

```python
url = 'https://www.lagaceta.com.ar/deportes'
```

```python
# path_to_extension = r'C:\Users\arzan\Desktop\1.17.0_0'
windows = 'chromedriver.exe'
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")
# chrome_options.add_argument('load-extension=' + path_to_extension)
chrome_options.add_argument("--start-maximized")
```

```python
# Id del contenedor = 'notas'
# Clase de cada nota = 'addItem'
# titulo en h2
# en h2 -> a[2] esta la url de la nota
# div class time esta la fecha
lagacetaDeportes = []

driver = webdriver.Chrome(windows, chrome_options=chrome_options)
driver.create_options()
driver.get(url)

# locator = (By.ID, 'icon_784515')
# EC.presence_of_element_located(locator)


maxScrol = .1
while maxScrol < 9.9:
    try:
        miBusqueda = driver.find_element_by_id('icon_784220')
        print(miBusqueda)
        maxScrol = 10
    except:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight/%s);" % maxScrol)
        maxScrol += .001


# maxScrol = 5
# driver.execute_script("window.scrollTo(0, document.body.scrollHeight/%s);" % maxScrol)

contenedor = driver.find_element_by_id("notas")

notas = contenedor.find_elements_by_class_name("card")
listOfWebElements = notas

# print(notas[4].find_element_by_class_name("text").text)

count = 0
for i in reversed(notas):
    nota_id = count
    titulo = i.find_element_by_tag_name("h2").text
    url = i.find_elements_by_tag_name("a")[1].get_attribute('href')
    
    try:
        autor = i.find_elements_by_tag_name("a")[2].text
    except:
        autor = 'No autor'
        
    fecha = i.find_element_by_class_name("time").text
    seccion = 'deportes'
    
    lagacetaDeportes.append({'nota_id':nota_id,'autor':autor,'fecha':fecha,'seccion':seccion,'titulo':titulo,'url':url})
    
    print('+++ ' , count , fecha)
    count+=1
    
# print (titulo)
driver.quit()
    
```
