### INSTANCIAS DE LA BUSQUEDA

```python
#PARAMETROS DE BUSQUEDA

#Intervalo de fecha de busqueda - Igualados se realiza a busqueda de un dia en particular.

dDia = hDia = 20 #desde dia hasta dia
dMes = hMes = 9 #desde mes hasta mes
dAno = hano = 2018 #desde año hasta año

#Termino de bsqueda

termino = '*'

#Sitio
protocolo = 'https'
url = 'www.clarin.com'

#Cantidad de resultados
cantR = '100'
```

```python
mydate = datetime.date(dAno,dMes, dDia) #Busqueda DESDE
fecha = mydate.strftime("%Y-%m-%d")
user_agent = {'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:59.0) Gecko/20100101 Firefox/59.0'}
googleUrl = 'https://www.google.com.ar/search?q='
```

## Captura de enlaces en las busqueda de Google

```python
source = []
count = 0
for dia in range(31):
# -> Seteo de fecha
    fecha = mydate.strftime("%Y-%m-%d")
    d = fecha.split('-')
    dDia = hDia = d[2]
    dMes = hMes = d[1].lstrip('0')
    dAno = hano = d[0]
    
# ->Contruccion de enlace a busqueda de Goolge     
    #query = termino + '+site%3A' + protocolo + '%3A%2F%2F' + url + '%2F&num=' + cantR + '&source=lnt&tbs=cdr%3A1%2Ccd_min%3A' + dMes + '%2F' + dDia + '%2F' + dAno + '%2Ccd_max%3A' + hMes + '%2F' + hDia + '%2F' + hano + '&tbm='
    query = termino + '+site%3A' + protocolo + '%3A%2F%2F' + url + '%2F&num=' + cantR + '&source=lnt&tbs=cdr%3A1%2Ccd_min%3A' + dMes + '%2F' + dDia + '%2F' + dAno + '%2Ccd_max%3A' + hMes + '%2F' + hDia + '%2F' + hano + '&tbm='
    consulta = googleUrl + query

# -> Recontruccion de la fecha en cuestion
    fecha = (dDia+'-'+dMes+'-'+dAno)
    
# -> Conexion a consulta de Google    
    response = requests.get(consulta, headers = user_agent)
# -> Soup de Google
    google_soup = bs(response.text, 'html.parser')
    
    print(response.status_code)
    print(consulta)

# -> Captura de enlaces resultantes de la busqueda de Google
    for d in google_soup.find_all("div", {"class": "r"}):
        res = d.a.attrs['href']
# -> COntruccion de la lista de enlaces con su fecha correspondiente        
        source.append((res,fecha))
        
    
    mydate += timedelta(days=1)
    count += 1
    print(count , res)
```

```python

# BARRIDO DE URL´S

sectionList = []
okList = []
count = 0

for url , fecha in source:

# ->
    print(url)
    
    response = requests.get(url, headers = user_agent)
    status = response.status_code
    
# -> Check de conexion
    print('status :',status)
    
    #soup
    soup = bs(response.text, 'html.parser')
    
# -> Check de Beautiful
    print(len(soup)>1)
    
    #Seccion incluida en la nota?? 
    try:
        titular= soup.find("div", {"class": "title"}).find("h1" , {"id":"title"}).text
    except:
        titular = 'sin titulo'
    try:
        subTitular= soup.find("div", {"class": "title"}).find("h2").text
    except:
        subTitular = 'sin subtitulo'
    try:
        autor = soup.find("div", {"class": "content-top"}).parent.find("p").text
    except:
        autor = 'sin autor'
    try:
        sectionIn = soup.find("div", {"class": "content-top"}).parent.find_all("li")[5].text.lower()
    except:
        setionIn = 'sin seccion'
        
        
# -> Check de secciones  
    check = sectionIn in main_category
    
    print( 'url :',url , 'status :',status , 'Seccion :' , sectionIn ,' ' , check ,'\n')
    
    if check:
        sectionList.append(sectionIn)
        
# -> LA NOTA
        
        try:
            contenedor = soup.find('div' ,{'class':'body-nota'})
            flag = 0
            for p in contenedor.find_all('p'):
                if flag:
                    nota = nota + p.text
                else:
                    nota = p.text
                    flag = 1
        except:
            continue

        okList.append({'titulo': titular,
                       'autor':autor,
                       'cuerpo':nota,
                       'url': url,
                       'fecha':fecha,
                       'seccion': sectionIn,
                       'abierta_comentarios': 'true'})
        
        
```
